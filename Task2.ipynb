```python
import re
def clean_text(text):
  text = re.sub(r'\s+', ' ', text)  # Replace multiple spaces/newlines with a single space
  text = re.sub(r'^\s+|\s+?$', '', text)  # Remove leading/trailing spaces
  text = re.sub(r'Page \d+', '', text) # Remove "Page X" type patterns
  text = re.sub(r'^[A-Za-z0-9,\s]+$', '', text)  # Remove author names or titles in the header
  return text
```

```python
import os
import pdfplumber

def read_pdfs_from_folder(folder_path):
    # Dictionary to store the extracted text
    extracted_info = {"CVPR": [], "EMNLP": [],"KDD": [],"NeurIPS": [],"TMLR": []}

    # Loop through each subfolder ('Publishable' and 'Non-Publishable')
    for subfolder in ['CVPR','EMNLP','KDD','NeurIPS','TMLR']:
        subfolder_path = os.path.join(folder_path, subfolder)

        if os.path.exists(subfolder_path):
            # Recursively walk through all directories and files inside the subfolder
            for root, dirs, files in os.walk(subfolder_path):
                pdf_files = [f for f in files if f.endswith('.pdf')]

                # Loop through each PDF file found in the current directory
                for pdf_file in pdf_files:
                    file_path = os.path.join(root, pdf_file)

                    # Open the PDF and extract the text
                    with pdfplumber.open(file_path) as pdf:
                        text = ""
                        for page in pdf.pages:
                            text += page.extract_text()

                        # Clean the extracted text before appending
                        extracted_info[subfolder].append(clean_text(text))
        else:
            print(f"No Folder named {subfolder}")

    return extracted_info
```

```python
import torch
import faiss
import numpy as np
from transformers import BertTokenizer, BertModel
from sklearn.metrics.pairwise import cosine_similarity
import math

# Step 1: Use BERT Tokenizer and BERT Model to Extract Features
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
model = BertModel.from_pretrained('bert-base-uncased')
def extract_keywords(text):
    # Tokenize the input text using BERT tokenizer
    inputs = tokenizer(text, return_tensors="pt", truncation=True, padding=True, max_length=12)

    # Pass the tokenized input through BERT to get the embeddings
    with torch.no_grad():
        outputs = model(**inputs)

    # Get the embeddings of the [CLS] token (first token) from the output hidden states
    embeddings = outputs.last_hidden_state[:, 0, :].numpy()  # Shape: (1, 768)

    return embeddings

# Step 2: Create FAISS Index and Add Conference Profiles
def create_faiss_index(conference_profiles):
    embeddings = []
    conference_names = []

    # Get the embeddings for each conference profile
    for conference, keywords in data.items():
        profile_text = ' '.join(keywords)  # Concatenate keywords for the conference
        embedding = extract_keywords(profile_text).flatten()
        embeddings.append(embedding)
        conference_names.append(conference)

    # Convert the list of embeddings into a numpy array
    embeddings = np.array(embeddings).astype('float32')

    # Initialize a FAISS index (using L2 distance)
    index = faiss.IndexFlatL2(embeddings.shape[1])  # L2 distance
    index.add(embeddings)  # Add the conference embeddings to the index

    return index, conference_names

# Step 3: Compare Paper to Conference Profile Using FAISS
def recommend_conference(paper_text, index, conference_names):
    paper_embedding = extract_keywords(paper_text).flatten().reshape(1, -1)

    # Search for the most similar conferences in the FAISS index
    distances, indices = index.search(paper_embedding, k=1)  # Top 3 most similar conferences

    recommendations = []
    for i, idx in enumerate(indices[0]):
        conference = conference_names[idx]
        similarity =1-math.sqrt(distances[0][i])  # Convert distance to similarity (1 - distance)
        recommendations.append((conference, similarity))

    return recommendations

# Step 4: Justification for Recommendations
def generate_justification(paper_text, conference_name, recommendations):
    # Find the corresponding similarity score from the recommendations
    similarity = next((similarity for conf, similarity in recommendations if conf == conference_name), 0)

    if similarity > 0.5:  # Threshold for good match
        return f"The paper aligns well with {conference_name}, focusing on similar research areas such as {', '.join(data[conference_name])}. The methodology and findings are consistent with those commonly presented at {conference_name}."
    else:
        return f"The paper's topic overlaps moderately with {conference_name}. However, it may be more suitable for other conferences focusing on broader themes."


# Step 5: Final Classification (Full Paper Classification Process)
def classify_paper(extracted_info):
    # Create the FAISS index with conference profiles
    index, conference_names = create_faiss_index(data)

    # Iterate through extracted papers in the Test folder
    for text in extracted_info["Test"]:
        paper_text = text

        # Recommend conferences based on the paper's content
        recommendations = recommend_conference(paper_text, index, conference_names)
        # Generate justifications for the top 3 conferences
        justification = {}
        for conference, _ in recommendations[:1]:
            dict2['Conference'].append(conference)
            justification[conference] = generate_justification(paper_text, conference,recommendations)
            dict2['Rationale'].append(justification[conference])

        # Print results
        print("Title:")
        print(text[:50])  # Print the first 50 characters as a preview
        print("." * 80)
        for conference, similarity in recommendations[:3]:
            print(f"Recommended Conference: {conference}, Similarity Score: {similarity}")
            print(f"Justification: {justification[conference]}")
            print("-" * 80)

    return recommendations[:3], justification


# Example Usage:
folder_path = "path/to/your/folder"
test_paper_folder = "Test_Papers"
paper_name = "your_paper_title_or_identifier"


# Run the classification
recommended_conferences, justification = classify_paper(extracted_info)

# Output the results
for conference, score in recommended_conferences:
    print(f"Recommended Conference: {conference}, Similarity Score: {score}")
    print(f"Justification: {justification[conference]}")
    print("-" * 80)
```
